```python
import pandas as pd
import numpy as np
```

```python
datapointsix = pd.read_csv('D:/aayush/minor/archive/data/GP1_0.6_marker.csv')
datapointseven = pd.read_csv('D:/aayush/minor/archive/data/GP1_0.7_marker.csv')

dataonepointone = pd.read_csv('D:/aayush/minor/archive/data/GP1_1.1_marker.csv')
dataonepointtwo = pd.read_csv('D:/aayush/minor/archive/data/GP1_1.2_marker.csv')

dataonepointsix = pd.read_csv('D:/aayush/minor/archive/data/GP1_1.6_marker.csv')
dataonepointseven = pd.read_csv('D:/aayush/minor/archive/data/GP1_1.7_marker.csv') 
```

```python
data0 = pd.DataFrame({'secim':np.zeros(12000)})
data1 = pd.DataFrame({'secim':np.ones(12000)})
class2 = []
class3 = []
class4 = []
class5 = []

for i in range(1,12001):
    class2.append(2)
    class3.append(3)
    class4.append(4)
    class5.append(5)

data2 = pd.DataFrame({'secim':class2})
data3 = pd.DataFrame({'secim':class3})
data4 = pd.DataFrame({'secim':class4})
data5 = pd.DataFrame({'secim':class5})
# print(data3)

data5 = pd.concat([dataonepointseven,data5],axis=1)
data4 = pd.concat([dataonepointsix,data4],axis=1)
data3 = pd.concat([dataonepointtwo,data3],axis=1)
data2 = pd.concat([dataonepointone,data2],axis=1)
data1 = pd.concat([datapointseven,data1],axis=1)
data0 = pd.concat([datapointsix,data0],axis=1)

data4 = data4.append(data5,ignore_index=True)
data3 = data3.append(data4,ignore_index=True)
data2 = data2.append(data3,ignore_index=True)
data1 = data1.append(data2,ignore_index=True)
data = data0.append(data1,ignore_index=True)

# data.head()
```

```python
x = data.drop(["secim"],axis=1)
y = data.secim.values
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print('Accuracy of dec tree in test data is:',dt.score(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100,random_state=1)
rf.fit(x_train,y_train)
print('Random Forest accuracy on test data is : ',rf.score(x_test,y_test))
```
